```python
import numpy as np
import matplotlib.pyplot as plt
import math
import tensorflow as tf

def step_math(x):
    return 1 if x >= 0 else 0

def step_tf(x):
    return tf.cast(tf.greater_equal(x, 0), tf.float32)

inputs = np.linspace(-10, 10, 100)
outputs_math = [step_math(x) for x in inputs]
outputs_tf = step_tf(inputs)

plt.figure(figsize=(8, 6))
plt.plot(inputs, outputs_math, label='Math Library', linestyle='--', color='blue')
plt.plot(inputs, outputs_tf, label='TensorFlow', linestyle='-', color='red')
plt.title('Step Function')
plt.xlabel('Input')
plt.ylabel('Output')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import math
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf

def relu_math(x):
    return max(0, x)

def relu_tf(x):
    return tf.nn.relu(x)

x_values = np.linspace(-5, 5, 100)

y_math = [relu_math(x) for x in x_values]
y_tf = [relu_tf(x) for x in x_values]

plt.plot(x_values, y_math, label='ReLU (Math Library)')
plt.plot(x_values, y_tf, label='ReLU (TensorFlow)')
plt.xlabel('Input')
plt.ylabel('Output')
plt.title('ReLU Activation Function')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
import math

def sigmoid_math(x):
    return 1 / (1 + math.exp(-x))

def sigmoid_tf(x):
    return tf.sigmoid(x)

x_values = np.linspace(-10, 10, 100)

y_math = [sigmoid_math(x) for x in x_values]
y_tf = sigmoid_tf(x_values)

plt.plot(x_values, y_math, label='Sigmoid (Math Library)')
plt.plot(x_values, y_tf, label='Sigmoid (TensorFlow)')
plt.xlabel('Input')
plt.ylabel('Output')
plt.title('Sigmoid Activation Function')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
import math

def leaky_relu_math(x, alpha=0.01):
    return max(alpha * x, x)

def leaky_relu_tf(x, alpha=0.01):
    return tf.nn.leaky_relu(x, alpha=alpha)

x_values = np.linspace(-5, 5, 100)

alpha_values = [0.01, 0.1, 0.5]

plt.figure(figsize=(10, 6))

for alpha in alpha_values:
    y_math = [leaky_relu_math(x, alpha) for x in x_values]
    y_tf = leaky_relu_tf(x_values, alpha)

    plt.plot(x_values, y_math, label=f'Leaky ReLU (Math Library) Alpha={alpha}')
    plt.plot(x_values, y_tf, label=f'Leaky ReLU (TensorFlow) Alpha={alpha}')

plt.xlabel('Input')
plt.ylabel('Output')
plt.title('Leaky ReLU Activation Function')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import tensorflow as tf

def softmax_math(x):
    exp_values = np.exp(x - np.max(x, axis=-1, keepdims=True))
    return exp_values / np.sum(exp_values, axis=-1, keepdims=True)

def softmax_tf(x):
    return tf.nn.softmax(x)

# Example usage:
x = np.array([2.0, 1.0, 0.1])
print("Softmax using math library:", softmax_math(x))
print("Softmax using TensorFlow:", softmax_tf(x))

```
